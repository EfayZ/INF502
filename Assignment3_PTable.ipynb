For the main code, I need to use the information from the table, and provide 5 options to users that they could operate with it. At the same time, you need to notice users that you can just input 1-5 to do the operation.

```python
import json

with open('PeriodicTableJSON.json','r') as PT:
        PTable = json.load(PT)
PT.close()
Elements = PTable['elements']

def PeriodicTable(Elements): 
    while 1:
        Options = input("1: See all the information that is stored about any element.\n \
2: Choose a property, and see that property for each element in the table.\n \
3: Enter a new element\n \
4:Change the attributes of an element\n \
5:Exit the program\n")
        try:
            Options = int(Options)
            if Options >5 or Options <1:
                raise ValueError
        except ValueError:
            print("You have to input a valid number!")
        else:
            if  Options == 1:
                Option1(Elements)
            elif Options == 2:
                Option2(Elements)
            elif Options == 3:
                Option3(Elements)
            elif Options == 4:
                Option4(Elements)
            elif Options == 5:
                return
```

For option1, I used the "flag" to jugde if the symbol is exsited in the table. If yes, then users will get the information.

```python
def Option1(Elements):
    
    Sign = 0
    while Sign == 0:
            Symbol = str(input("Enter the element's symbol\n"))
            for Element in Elements:
                if Symbol == Element["symbol"]:
                    Sign = 1
                    break
            if Sign == 0:
                print("Check if your symbol is exsited!")
            else:
                print("symbol : %s" %Element["symbol"])
                print("name : %s" %Element["name"])
                print("number : %s" %Element["number"])
                print("row : %s" %Element["xpos"])
                print("column : %s" %Element["ypos"])
            
        
```

For option2, I typed out everything that included properties. Compare if the input contained the properties that listed.If yes, users will get all the related information.

```python
def Option2(Elements):
    Properties = {"symbol", "name", "number", "xpos", "ypos"}
    
    Sign = 0
    while Sign == 0:
            Property = str(input("Enter the property:\n"))
            for Subproperty in Properties:
                if Property == Subproperty:
                    Sign = 1
                    break
            if Sign == 0:
                    print("Check if your input property is right!\n")  
            else:
                for Element in Elements:
                    print("Element:%s %s : %s" %(Element["symbol"], Property, Element[Property]))
```

For option 3, users will input every property they knew. If there is something that system knew, so the element is exsited. Or it will add to the table. 

```python
def Option3(Elements):
    Sign1 = 0
    while Sign1 == 0:
        Sign2 = 0
        NewSymbol = str(input("Please input a new element symbol:\n"))
        NewName = str(input("Please input a new element name:\n"))
        NewNumber = str(input("Please input a new element number:\n"))
        NewX = str(input("Please input a new element xpos:\n"))
        NewY = str(input("Please input a new element ypos:\n"))
        for Element in Elements:
            if NewSymbol == Element["symbol"] or NewName == Element["name"] or NewName == Element["number"]:
                Sign2 = 1
                break
            else:
                Sign1 = 1               
        if Sign1 == 0:
            print("Your new element is exsited in the table.\n")
        else:
            Sign1 = 1
            NewMember= {"symbol":NewSymbol, "name":NewName, "number":NewNumber, "xpos":NewX,\
                     "ypos":NewY}
            print("Your new element:",NewMember)
            Elements.append(NewMember)
    return Elements
```

For option4, I need to judge if the elements exsit. If the element exsited, then users could make changes they want. 

```python
def Option4(Elements):
    
    Sign1 = 0
    while Sign1 == 0:
        Symbol = str(input("Enter the element's symbol\n"))
        for Element in Elements:
            if Symbol == Element["symbol"]:
                    Sign1 = 1
                    break
            else:
                print("Your symbol is not exsited.\n")
        if Sign1 == 1:
            print("Please make your changes in attributes!\n")
            NewSymbol = str(input("Please input a new element symbol:\n"))
            NewName = str(input("Please input a new element name:\n"))
            NewNumber = str(input("Please input a new element number:\n"))
            NewX = str(input("Please input a new element xpos:\n"))
            NewY = str(input("Please input a new element ypos:\n"))
    
            Element["name"] = NewName

            Element["number"] = NewNumber

            Element["xpos"] = NewX

            Element["ypos"] = NewY
            print("After changing:",Element)
    return Elements
    
```
